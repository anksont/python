```python
di={"name of laptop":[],"price":[],"discount":[],"rating and reviews":[],"emi":[]}
for i in range(len(p)):
    n=page.findAll("div",{"class":"_3wU53n"})
    name=n[i].text
    di["name of laptop"].append(name)
    p=page.findAll("div",{"class":"_6BWGkk"})
    #print(p[i].find("div",{"class":"_1vC4OE _2rQ-NK"}).text)
    original_price=p[i].find("div",{"class":"_1vC4OE _2rQ-NK"}).text
    di["price"].append(original_price)
    #print(p[i].find("div",{"class":"VGWI6T"}).text)
    discount=p[i].find("div",{"class":"VGWI6T"}).text
    di["discount"].append(discount)
    r=page.findAll("div",{"class":"niH0FQ"})
    #rating=r[1].find("div",{"class":"hGSR34"}).text
    rating_and_reviews=r[i].find("span",{"class":"_38sUEc"}).text
    #print(rating_and_reviews)
    di["rating and reviews"].append(rating_and_reviews)
    e=page.findAll("div",{"class":"_3MCpsc"})
    emi=e[i].text.split()
    di["emi"].append(emi[3])    

di
df=pd.DataFrame(di)
    
```
