```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import os
os.getcwd()
```

```python
xlsx=pd.ExcelFile("C:\\Users\\Pranay\\Desktop\\rakesh data and analysis\\Reservoir-2018.xlsx",index_col=None)
df=pd.read_excel(xlsx,"TB-2018")
```

```python
df.set_index("Flow Date",drop=True,inplace=True)
```

```python
df_pm=pd.DataFrame((df.loc[df.index <="2018-5-31"]).mean().T,index=None)
```

```python
df_pm[1]=pd.DataFrame((df.loc[(df.index<="2018-9-30") & (df.index >"2018-5-31")]).mean().T,index=None)
```

```python
df_pm[2]=pd.DataFrame((df.loc[(df.index<="2018-12-31") & (df.index >"2018-9-30")]).mean().T,index=None)
```

```python
df_pm[3]=pd.DataFrame((df.loc[df.index <="2017-5-31"]).mean().T,index=None)
df_pm[4]=pd.DataFrame((df.loc[(df.index<="2017-9-30") & (df.index >"2017-5-31")]).mean().T,index=None)
df_pm[5]=pd.DataFrame((df.loc[(df.index<="2017-12-31") & (df.index >"2017-9-30")]).mean().T,index=None)
```

```python
df_pm.columns=['PM-2018','SWM-2018','NEM-2018','PM-2017','SWM-2017','NEM-2017']
```

```python
dfm=df_pm.T.copy()
```

```python
dfm.plot.bar(rot=90,subplots=True,layout=(2,4),figsize=(12,10),colormap="plasma",sharex=True,use_index=True,legend=False)
plt.tight_layout()
plt.savefig("reservior visualisation.pdf")

```
