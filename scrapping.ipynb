```python
url="https://www.amazon.com/s?k=shoe&ref=nb_sb_noss_2"
```

```python
#connection to the web page
uconn=uReq(url)
page_content=uconn.read()
uconn.close()
```

```python
page=soup(page_content,"html.parser")
```

```python
print(page.prettify())
```

```python
containers=page.findAll("div",{"class" : "s-expand-height s-include-content-margin s-border-bottom"})
```

```python
len(containers)
```

```python
print(soup.prettify(containers[40]))
container=containers[40]
i=container.find("div",{"class":"a-section aok-relative s-image-tall-aspect"})
sneaker_name=i.img["alt"]
print(sneaker_name)
try:
    r=container.find("div",{"class":"a-row a-size-small"})
    rating=r.span.text.strip()
    print(rating)
except:
    rating="None"
    print(rating)
    
try:
    cr=container.find("span",{"class":"a-size-base"})
    customer_reviews=cr.text
    print(customer_reviews)
except:
    customer_reviews="none"
    print(customer_reviews)
try:                                         
    p=container.findAll("span",{"class":"a-offscreen"})
    price_range=str(p[0].text)+"-"+str(p[1].text)
    print(price_range)
except:
    price_range=str(p[0].text)+"- NM"
    print(price_range)
```

```python
import numpy as np
x=np.arange(8,103,1)
import pandas as pd
```

```python
di={"sneaker name":[],"sneaker rating":[],"customer reviewed":[],"price":[]}
for  container in containers :
    #container=containers[2]
    i=container.find("div",{"class":"a-section aok-relative s-image-tall-aspect"})
    sneaker_name=i.img["alt"]
    di["sneaker name"].append(sneaker_name)
    try:
        r=container.find("div",{"class":"a-row a-size-small"})
        rating=r.span.text.strip()
        di["sneaker rating"].append(rating)
    except:
        rating="None"
        di["sneaker rating"].append(rating)
        
    try:
        cr=container.find("span",{"class":"a-size-base"})
        customer_reviews=cr.text
        di["customer reviewed"].append(customer_reviews)
    except:
        customer_reviews="none"
        di["customer reviewed"].append(customer_reviews)
        
    try:                                         
        p=container.findAll("span",{"class":"a-offscreen"})
        price_range=str(p[0].text)+"-"+str(p[1].text)
        di["price"].append(price_range)
    except:
        price_range=str(p[0].text)+"- NM"
        di["price"].append(price_range)
 
```

```python
di
df=pd.DataFrame(di)
```

```python
df.columns=['sneaker name'.upper(), 'sneaker rating'.upper(), 'customer reviewed'.upper(), 'price range'.upper()]
```
